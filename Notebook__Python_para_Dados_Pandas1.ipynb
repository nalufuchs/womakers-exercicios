```python
df_concurso = pd.read_csv('dados_concurso.csv')
```

```python
df_concurso.head()
```

Qual a média de idade por estado?

```python
df_media_idade_estado = df_concurso[['Estado', 'Escolaridade', 'Número de Inscrição']].groupby(['Estado', 'Escolaridade']).count()

df_media_idade_estado.head(10)
```

Qual é a porcentagem de pessoas com deficiência?

```python
df_concurso['Deficiência'].value_counts(normalize=True)
```

ou:

```python
total_def = df_concurso['Deficiência'].count()
total_def
```

```python
df_concurso_deficiencia = (df_concurso['Deficiência'] == 'Sim').sum()
df_concurso_deficiencia
```

```python
df_concurso_porcentagem = df_concurso_deficiencia * 100 / total_def
df_concurso_porcentagem
```

ou:

```python
((df_concurso['Deficiência'] == 'Sim').sum()) * 100 / df_concurso['Deficiência'].count()
```

Quantas pessoas são do estado do Piauí?

```python
df_insc_por_estado = df_concurso[['Número de Inscrição', 'Estado']].groupby('Estado').count()
df_insc_por_estado
```

```python
(df_concurso['Estado'] == 'PI').sum()
```
