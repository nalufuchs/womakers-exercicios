```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("collected research people.csv")
```

```python
df.head()
```

Percebe-se que a coluna "Unnamed" é desnecessária à tabela, portanto, deleta-se.

```python
df = df.drop(columns=['Unnamed: 0'])
df.head()
```

Percebe-se nomes de colunas inconsistentes e com erros, portanto, padroniza-se com o comando abaixo:

```python
novas_cols = {'Endereerço': 'Endereço',
              'phone': 'Telefone',
              'anual_income': 'Renda Anual',
              'AAge': 'Idade',
              'n_filhos': 'N Filhos',
              'diariamente_sun blocker': 'Usa protetor solar diariamente',
              'Food_restrictions': 'Restrição alimentar',
              'streaming': 'Streaming'
              }

df.rename(columns=novas_cols, inplace=True)
df.head()
```

Observa-se, também, que os nomes da coluna Nome encontram-se com caracteres especiais. Corrige-se, sabendo que:

O **\W** equivale a [^a-zA-Zà-úÀ-Ú0-9_].

```python
df.Nome.str.replace('\W', '')
```

Porém, ao utilizar a expressão "\W", os espaços somem, bem como underlines permanecem, sem serem desejados. Assim, utilizando a equivalência da expressão regular, adiciona-se espaço e remove-se o "_" da expressão, obtendo, assim, uma coluna mais padronizada:

```python
df.Nome = df.Nome.str.replace('[^a-zA-Zà-úÀ-Ú0-9 ]', '')
```

```python
df.head()
```

Na coluna "Telefone", existe o mesmo problema, porém nem todos os caracteres especiais podem ser excluidos. Assim, separa-se os itens dentro do colchetes que irão permanecer na coluna, tais como os parênteses do DDD e o + do indicador de país, e, àqueles que não possuem os caracteres necessários para inclusão, terão "*" em sua composição, facilitando a detecção e exclusão de dados errôneos ou incorretos.

```python
df.Telefone = df.Telefone.str.replace('[^0-9() +-]', '*')
df.Telefone
```

Com esse comando, será verificado todos os números inválidos (aqueles que possuem "*"), e transformados em sentença booleana - para, assim, facilitar a visualização e permitir que a coluna possua números válidos ou NaN.

```python
df.Telefone.str.contains('[*]', na=False)
```

Assim, todos os números que possuíam espaços ou - permanecem, e àqueles incompletos possuirão "*", sendo de maior utilizade na exclusão dos números e deteccção de erros. Transforma-se em loc para facilitar manipulação e iguala-se a np.NaN do Numpy (para isso, todos os valores devem estar em booleano, conforme feito acima com na=False) e os que possuírem "*" em true (já que os NaN não necessitam serem alterados, recebem false como argumento)

```python
df.loc[df.Telefone.str.contains('[*]', na=False), 'Telefone'] = np.nan
```

```python
df.Telefone
```

Cenário 2: Uma empresa chamada Hillo, é uma empresa que está estudando o mercado e quer encontrar uma parceria com uma empresa de Streeming (netflix, Disney plus etc), mas gostaria de saber quais as empresas dariam o maior retorno de investimento.
Eles podem fazer análise de machine learning também com esses dados.


```python
df_hillo = df.copy()
```

```python
df_hillo.drop(columns=['Moradia',
                  'Restrição alimentar',
                  'Usa protetor solar diariamente',
                  'Vai na Praia Mensalmente',
                  'Faz Academia',
                  'Compra e Lê Livros Todos os Anos',
                  'Quantidade de Livros Comprados',
                  'Autor Favorito'], inplace=True)
```

Como o objetivo é streaming, observa-se a categoria para possível ordenação (separar quando existir mais de uma categoria selecionada, etc)

```python
df_hillo.Streaming
```

Existe, como observado, opções como "Netflix E disney" ou "Netflix E Hbo" que, para organizar melhor, separa-se.

```python
df_hillo.Streaming = df_hillo.Streaming.replace({'Netflix e Disney' : 'Netflix e Disney Plus'})
```

```python
df_hillo.Streaming
```

```python
df_hillo.Streaming = df_hillo.Streaming.replace({'Todos' : 'Globo Play e Netflix e HBO e Disney Plus e Crunchyroll'})
```

```python
df_hillo_stream = df_hillo['Streaming'].str.get_dummies(sep=' e ')
df_hillo_stream
```

Agora, com os valores separados, é possível adicioná-los novamente à tabela utilizando concat

```python
df_hillo = pd.concat([df_hillo, df_hillo_stream], axis=1)
```
